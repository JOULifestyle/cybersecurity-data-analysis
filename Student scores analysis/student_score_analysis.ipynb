# Student Academic Performance Analysis with Pandas.

```python
# import pandas
import pandas as pd
```

```python
# load dataset
url = 'https://raw.githubusercontent.com/ritaafrica/data/refs/heads/main/student_scores.csv'
df = pd.read_csv(url)
```

```python
df.head()
```

```python
# Adding a new column for total scrores
df['Total_Score'] = df['Math_Score'] + df['Reading_Score'] + df['Writing_Score']
```

```python
df.head()
```

```python
# Adding new column to calculate %score rounding it up to 2 decimal places
df['%Score'] = (df['Total_Score']/300*100).round(2)
```

```python
df.tail()
```

```python
# Adding new column to display pass or fail based on the % score
df['Status'] = df['%Score'].apply(lambda x: 'Pass' if x>=40 else 'Fail')
```

```python
df.head
```

```python
# Replacing the 'Math_Score' column to 'Mathematics_Score'
df.rename(columns={'Math_Score': 'Mathematics_Score'}, inplace=True)
```

```python
df.head()
```

```python
# Remove the 'School' column from the DataFrame
df.drop(columns={'School'}, inplace=True)
```

```python
df.head()
```

```python
# Saving dataset to csv
df.to_csv('student_scores_sheet.csv', index=False)
```

```python
# Saving dataset to html
df.to_html('student_scores_sheet.html', index=False)
```
